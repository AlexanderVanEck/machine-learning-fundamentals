```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plot
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset = pd.read_csv('datasets/position_salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor = regressor.fit(X, y)
```

```python
# the DecisionTreeRegressor takes the average between two data points as the predicted
# value, this means the output plot should be a blocky line shooting up where the data point
# changes to a different average. This is called a non-continuous regression model.
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plot.scatter(X, y, color='red')
plot.plot(X_grid, regressor.predict(X_grid), color='blue')
plot.xlabel('Position level')
plot.ylabel('Yearly salary')
plot.title('Position vs. salary')
plot.show()
```
