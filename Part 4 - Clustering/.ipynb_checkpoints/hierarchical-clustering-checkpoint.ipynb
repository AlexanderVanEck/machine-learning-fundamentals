```python
import matplotlib.pyplot as plot
import numpy as np
import pandas as pd
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
```

```python
dataset = pd.read_csv('datasets/customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
# With the hierachical clustering algorithm we are going to cluster
# our dataset into multiple clusters. First we have to find the optimal
# amount of clusters. This can be done with a dendrogram.

# the ward method tries to minimise the varience in the cluster
dendogram = hierarchy.dendrogram(hierarchy.linkage(X, method='ward'))
plot.title('Dendogram')
plot.xlabel('Customers')
plot.ylabel('Euclidean distances')
plot.show()
```

```python
# from the denogram above we can see that the longest distance without
# crossing a horizontal line is the right most vertical line.
# A horizontal line that crosses this vertical line will cross 5 lines,
# therefore our optimal amount of clusters is 5.

# now we fit a AgglomertiveClustering algorithm to cluster our data.
clusterer = AgglomerativeClustering(
    n_clusters=5,
    affinity='euclidean',
    linkage='ward'
)
y_clusters = clusterer.fit_predict(X)
```

```python
# Visualising each cluster
labels = [
    'Careful',
    'Standard',
    'Target',
    'Careless',
    'Sensible',
]
for i in range(5):
    plot.scatter(
        X[y_clusters == i, 0], X[y_clusters == i, 1],
        label=labels[i]
    )
plot.title('Clusters of customers')
plot.xlabel('Annual Income (k$)')
plot.ylabel('Spending Score (1..100)')
plot.legend()
plot.show()
```
